```python
def getcasegendernumber(text, nlp_type):
    '''получает падеж род число из NEs'''
    featscombo = []
    nes = re.findall(r'\[(.*?)\]', text)

    for ne in nes:
        doc = nlp_type(ne)
        for sent in doc.sentences:
            for word in sent.words:
                if word.feats:
                    parts = []
                    for feat in word.feats.split('|'):
                        if feat.startswith(('Case=', 'Gender=', 'Number=')):
                            parts.append(feat)
                    if parts:
                        featscombo.append('|'.join(parts))

    return featscombo
```

```python
fillers = {
    'be': {
        'PER': {'Masc': {'Sing': 'мужчына', 'Pl': 'мужчыны'},
                'Fem': {'Sing': 'жанчына', 'Pl': 'жанчыны'},
                'Neut': {'Sing': 'стварэнне', 'Pl': 'стварэнні'}},
        'LOC': {'Masc': {'Sing': 'горад', 'Pl': 'гарады'},
                'Fem': {'Sing': 'вёска', 'Pl': 'вёскі'},
                'Neut': {'Sing': 'паселішча', 'Pl': 'паселішчы'}},
        'ORG': {'Masc': {'Sing': 'саюз', 'Pl': 'саюзы'},
                'Fem': {'Sing': 'арганізацыя', 'Pl': 'арганізацыі'},
                'Neut': {'Sing': 'вытворчасць', 'Pl': 'вытворчасці'}}
    },
    'sl': {
        'PER': {'Masc': {'Sing': 'moški', 'Pl': 'moški'},
                'Fem': {'Sing': 'ženska', 'Pl': 'ženske'},
                'Neut': {'Sing': 'bitje', 'Pl': 'bitja'}},
        'LOC': {'Masc': {'Sing': 'kraj', 'Pl': 'kraji'},
                'Fem': {'Sing': 'vas', 'Pl': 'vasi'},
                'Neut': {'Sing': 'naselje', 'Pl': 'naselja'}},
        'ORG': {'Masc': {'Sing': 'zaveznik', 'Pl': 'zavezniki'},
                'Fem': {'Sing': 'organizacija', 'Pl': 'organizacije'},
                'Neut': {'Sing': 'ustvarjanje', 'Pl': 'ustvarjanja'}}
    },
    'sk': {
        'PER': {'Masc': {'Sing': 'muž', 'Pl': 'muži'},
                'Fem': {'Sing': 'žena', 'Pl': 'ženy'},
                'Neut': {'Sing': 'stvorenie', 'Pl': 'stvorenia'}},
        'LOC': {'Masc': {'Sing': 'hrad', 'Pl': 'hrady'},
                'Fem': {'Sing': 'dedina', 'Pl': 'dediny'},
                'Neut': {'Sing': 'obydlie', 'Pl': 'obydlia'}},
        'ORG': {'Masc': {'Sing': 'zväz', 'Pl': 'zväzy'},
                'Fem': {'Sing': 'organizácia', 'Pl': 'organizácie'},
                'Neut': {'Sing': 'strojárstvo', 'Pl': 'strojárstva'}}
    }
}
```

```python
columninfo = {
    'be': {'text': 'translated_belarusian', 'feats': 'be_feats'},
    'sl': {'text': 'translated_slovenian', 'feats': 'sl_feats'},
    'sk': {'text': 'translated_slovak', 'feats': 'sk_feats'},
}
```

```python
def extractfeatures(feat_list):
    '''i extract features'''
    if not isinstance(feat_list, list):
        try:
            feat_list = eval(feat_list)  # преобразовать 
        except:
            return None
    for feat in feat_list:
        if 'Case=Nom' in feat:
            gender_match = re.search(r'Gender=(Masc|Fem|Neut)', feat)
            number_match = re.search(r'Number=(Sing|Plur)', feat)
            if gender_match and number_match:
                gender = gender_match.group(1)
                number = number_match.group(1)
                return gender, number
    return None
```

```python
def entityfinder(text):
    match = re.search(r'\[([^\[\]]+)\]', str(text))
    return match.group(1) if match else None
```

```python
def addfillers(input_path, output_path):
    '''меняет исходные сущности на филллеры'''
    df = pd.read_csv(input_path)

    for idx, row in df.iterrows():
        entity_type = row['ner_type']  # PER, ORG, LOC
        foundfeats = None
        for lang in columninfo:
            feats_col = columninfo[lang]['feats']
            feats = extractfeatures(row[feats_col])
            if feats:
                foundfeats = feats
                break

        if not foundfeats:
            continue  # нет features

        default_gender, default_number = foundfeats

        for lang, config in columninfo.items():
            text_col = config['text']
            feats_col = config['feats']
            text = row[text_col]

            if pd.isna(text): continue

            entitybr = entityfinder(text)
            if not entitybr: continue

            feats = extractfeatures(row[feats_col])
            gender, number = feats if feats else (default_gender, default_number)

            try:
                filler = fillers[lang][entity_type][gender][number]
            except KeyError:
                continue

            # []
            new_text = re.sub(r'\[' + re.escape(entitybr) + r'\]', f'[{filler}]', text, count=1)
            df.at[idx, text_col] = new_text

    df.to_csv(output_path, index=False)
    #print('файл сохранён')
```

```python
addfillers('templates_with_feats.csv', 'templates_fillers_added.csv')
```

```python
for col in ['be_feats', 'sl_feats', 'sk_feats']:
    data[col] = data[col].apply(lambda x: ast.literal_eval(x) if pd.notna(x) and isinstance(x, str) else [])
```

```python
# строки, где хотя бы в одной из колонок feats падеж не Nom
def nonom(feats_list):
    return any('Case=' in feat and 'Case=Nom' not in feat for feat in feats_list)
```

```python
# - те где в любом языке падеж не Nom
data_filtered = data[~(
    data['be_feats'].apply(nonom) |
    data['sl_feats'].apply(nonom) |
    data['sk_feats'].apply(nonom)
)]
```
